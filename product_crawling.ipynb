```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import os
```

```python
url = "https://www.jigushop.co.kr/all"
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

base_url = url
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)


```

```python
driver.get(base_url)
```

```python
html = driver.page_source #request.text대신 사용
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
item_list = soup.select('.shop-item._shop_item .item-pay')
print(len(item_list))
```

```python
image_list = soup.select('.shop-item._shop_item img')
print(len(image_list))
```

```python
items = soup.select('.shop-item._shop_item')
print(len(items))
```

```python
items
```

```python
image=[]
price=[]
name=[]
detail=[]
df = pd.DataFrame()

```

```python
for item in items:
  
  print("image : ",item.select("img")[0].get("src"))
  print("price : ",item.select("p")[0].text)
  print("name : ",item.select("h2")[0].text)
  detail_ = "https://www.jigushop.co.kr/"+item.select("a")[0].get('href')
  print("detail : ",detail_)

  image.append(item.select("img")[0].get("src"))
  price.append(item.select("p")[0].text)
  name.append(item.select("h2")[0].text)
  detail.append(detail_)


df['image'] = image
df['price'] = price
df['name'] = name
df['detail'] = detail
```

```python
df
```

```python
df.to_csv('item.csv',index=False, encoding="utf-8-sig")
```
